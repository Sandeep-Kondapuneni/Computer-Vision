# Canny edge detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image from file
image_path = 'musk babai.png'  # Replace with the actual path to your image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur to reduce noise and improve edge detection
blurred_image = cv2.GaussianBlur(original_image, (5, 5), 0)

# Apply Canny edge detector
edges = cv2.Canny(blurred_image, 50, 150)  # You can adjust the threshold values

# Display the original image, blurred image, and edges
plt.subplot(131), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(132), plt.imshow(blurred_image, cmap='gray'), plt.title('Blurred Image')
plt.subplot(133), plt.imshow(edges, cmap='gray'), plt.title('Edges')
plt.show()
```

# Sobel edge detection

```python
# sobel edge detection
image_path = 'musk babai.png'  # Replace with the actual path to your image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Sobel operator
sobel_x = cv2.Sobel(original_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(original_image, cv2.CV_64F, 0, 1, ksize=3)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalize the gradient magnitude to the range [0, 255]
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to uint8 for display
gradient_magnitude = np.uint8(gradient_magnitude)

# Display the original image, Sobel x, Sobel y, and gradient magnitude
plt.subplot(141), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(142), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel X')
plt.subplot(143), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel Y')
plt.subplot(144), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Gradient Magnitude')
plt.show()
```

# Scharr edge detection

```python
#scharr edge detection technique
image_path = 'musk babai.png'  # Replace with the actual path to your image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Scharr operator
scharr_x = cv2.Scharr(original_image, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(original_image, cv2.CV_64F, 0, 1)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(scharr_x**2 + scharr_y**2)

# Normalize the gradient magnitude to the range [0, 255]
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to uint8 for display
gradient_magnitude = np.uint8(gradient_magnitude)

# Display the original image, Scharr x, Scharr y, and gradient magnitude
plt.subplot(141), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(142), plt.imshow(scharr_x, cmap='gray'), plt.title('Scharr X')
plt.subplot(143), plt.imshow(scharr_y, cmap='gray'), plt.title('Scharr Y')
plt.subplot(144), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Gradient Magnitude')
plt.show()
```

# Laplacian of Gaussian (LoG)

```python
#Laplacian of Gaussian (LoG)
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur to the image
blurred_image = cv2.GaussianBlur(original_image, (5, 5), 0)

# Apply Laplacian of Gaussian (LoG)
log_image = cv2.Laplacian(blurred_image, cv2.CV_64F)

# Normalize the result to the range [0, 255]
log_image = cv2.normalize(log_image, None, 0, 255, cv2.NORM_MINMAX)

# Convert to uint8 for display
log_image = np.uint8(log_image)

# Display the original image, blurred image, and LoG result
plt.subplot(131), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(132), plt.imshow(blurred_image, cmap='gray'), plt.title('Blurred Image')
plt.subplot(133), plt.imshow(log_image, cmap='gray'), plt.title('LoG Edge Detection')
plt.show()
```

# Prewitt Operator

```python
# Prewitt Operator
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image from file
image_path = 'musk babai.png'  # Replace with the actual path to your image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Prewitt operator
prewitt_x = cv2.filter2D(original_image, cv2.CV_64F, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(original_image, cv2.CV_64F, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(prewitt_x**2 + prewitt_y**2)

# Normalize the gradient magnitude to the range [0, 255]
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to uint8 for display
gradient_magnitude = np.uint8(gradient_magnitude)

# Display the original image, Prewitt x, Prewitt y, and gradient magnitude
plt.subplot(141), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(142), plt.imshow(prewitt_x, cmap='gray'), plt.title('Prewitt X')
plt.subplot(143), plt.imshow(prewitt_y, cmap='gray'), plt.title('Prewitt Y')
plt.subplot(144), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Gradient Magnitude')
plt.show()

```

# Zero Crossing Edges

```python
from scipy.ndimage import convolve
# Apply Gaussian blur to the image
blurred_image = cv2.GaussianBlur(original_image, (5, 5), 0)

# Apply Laplacian of Gaussian (LoG)
log_image = cv2.Laplacian(blurred_image, cv2.CV_64F)

# Find zero crossings using convolution with a 3x3 kernel
kernel = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]])
zero_crossings = convolve(log_image, kernel)

# Threshold zero crossings
threshold = 0.01
edges = np.where(np.abs(zero_crossings) > threshold, 255, 0)

# Convert to uint8 for display
edges = np.uint8(edges)

# Display the original image, blurred image, LoG result, and zero crossing edges
plt.subplot(143), plt.imshow(log_image, cmap='gray'), plt.title('LoG Edge Detection')
plt.subplot(144), plt.imshow(edges, cmap='gray'), plt.title('Zero Crossing Edges')
plt.show()
```

# Robert Cross edge detection

```python
# Define the Robert Cross kernels
roberts_x = np.array([[1, 0], [0, -1]])
roberts_y = np.array([[0, 1], [-1, 0]])

# Apply the Robert Cross operators
roberts_x_edges = cv2.filter2D(original_image, cv2.CV_64F, roberts_x)
roberts_y_edges = cv2.filter2D(original_image, cv2.CV_64F, roberts_y)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(roberts_x_edges**2 + roberts_y_edges**2)

# Normalize the gradient magnitude to the range [0, 255]
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to uint8 for display
gradient_magnitude = np.uint8(gradient_magnitude)

# Display the original image, Robert Cross x edges, Robert Cross y edges, and gradient magnitude
plt.subplot(141), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(142), plt.imshow(roberts_x_edges, cmap='gray'), plt.title('Robert Cross X Edges')
plt.subplot(143), plt.imshow(roberts_y_edges, cmap='gray'), plt.title('Robert Cross Y Edges')
plt.subplot(144), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Gradient Magnitude')
plt.show()
```
